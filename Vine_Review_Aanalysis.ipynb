```python
import numpy
import pandas as pd
import warnings 

data= 'vine_table.csv'
warnings.filterwarnings('ignore')
```

```python
vine_df = pd.read_csv(data, index_col= False)
vine_df
```

```python
#filter total votes >= 20
filtered_df = vine_df[vine_df["total_votes"]>=20]
filtered_df
```

```python
#Filter the new DataFrame or table created in Step 1 and create a new DataFrame or 
#table to retrieve all the rows where the number of helpful_votes divided by 
#total_votes is equal to or greater than 50%.

filtered_df["ratio"] = filtered_df.helpful_votes/ filtered_df.total_votes >= .5
filtered_df
```

```python
vine_paid = filtered_df[filtered_df['verified_purchase']== 'Y']
vine_paid
```

```python
vine_unpaid = filtered_df[filtered_df['verified_purchase'] != "Y"]
vine_unpaid
```

```python
# Determine the total number of reviews, the number of 5-star reviews,
# and the percentage of 5-star reviews for the two types of review

vine_count = len(vine_paid)
nonvine_count= len(vine_unpaid)

vine_stars = (vine_paid['star_rating'].values == 5).sum()
nonvine_stars= (vine_unpaid['star_rating'].values == 5).sum()

vine_perc = (vine_stars/vine_count)
nonvine_perc= (nonvine_stars/nonvine_count)
```

```python
print(vine_count, vine_stars, vine_perc)
print(nonvine_count, nonvine_stars, nonvine_perc)
```
